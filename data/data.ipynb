```python
import pandas as pd
import xarray as xr
```

```python
ds = xr.open_dataset("climate_land_months_compressed.nc")

da = ds["temperature"].transpose("time", "latitude", "longitude")

france_mean = da.mean(dim=("latitude", "longitude"))
years = france_mean["time"].values
series = pd.Series(france_mean.values, index=years)
series.head()
series.to_csv("data.csv")
```
